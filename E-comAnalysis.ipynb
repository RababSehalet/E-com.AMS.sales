# Data Wrangling Process

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
EcomData = pd.read_csv('Amazon Sale Report.csv')
EcomData.head()
```

```python
EcomData.info()
```

```python
#Remove the data that won't be used in creating our charts
EcomData.drop(['currency', 'ship-postal-code', 'Unnamed: 22', 'SKU', 'Style'], axis=1, inplace=True)
EcomData.head()
```

```python
EcomData['promotion-ids'].value_counts()
```

```python
EcomData.drop(['ASIN'], axis=1, inplace=True)
EcomData.shape
```

```python
missing_values = EcomData.isnull().sum()
print(missing_values)
```

# Handling the missing Values

```python
# Define columns that should be filled with 'Unknown' (categorical)
categorical_columns = ['Courier Status', 'ship-city', 'ship-state', 'ship-country', 'promotion-ids']

# Replace missing values in categorical columns with 'Unknown'
EcomData[categorical_columns] = EcomData[categorical_columns].fillna('Unknown')

# Define numeric columns to fill with NaN (optional, as pandas automatically uses NaN for numeric missing values)
numeric_columns = ['Qty', 'Amount', 'B2B']

# Replace missing values in numeric columns with NaN
EcomData[numeric_columns] = EcomData[numeric_columns].apply(pd.to_numeric, errors='coerce')
# Fill missing numeric values with the median
EcomData[numeric_columns] = EcomData[numeric_columns].fillna(EcomData[numeric_columns].median())
```

```python
EcomData.info()
```

```python
EcomData.drop(['fulfilled-by'], axis=1, inplace=True)
```

```python
EcomData.tail()
```

```python
import plotly.express as px

# Group by city and calculate the total orders and sales (Amount)
city_data = EcomData.groupby(['ship-city']).agg({'Order ID': 'count', 'Amount': 'sum'}).reset_index()

# Map visualization (top cities by sales)
fig = px.choropleth(city_data, 
                    locations="ship-city", 
                    locationmode='country names',
                    color="Amount", 
                    hover_name="ship-city",
                    color_continuous_scale=px.colors.sequential.Plasma,
                    labels={"Amount": "Total Sales"})
fig.show()
```

```python
# Group by category and calculate total orders and sales (Amount)
category_data = EcomData.groupby(['Category']).agg({'Order ID': 'count', 'Amount': 'sum'}).reset_index()

# Top categories by number of orders
fig = px.bar(category_data, 
             x='Category', 
             y='Order ID', 
             title="Top Categories by Number of Orders",
             labels={"Order ID": "Number of Orders"})
fig.show()
```

```python

# Top categories by revenue
fig = px.bar(category_data, 
             x='Category', 
             y='Amount', 
             title="Top Categories by Revenue",
             labels={"Amount": "Total Sales"})
fig.show()
```

```python
# Group by shipping method and calculate total orders and sales (Amount)
shipping_data = EcomData.groupby(['ship-service-level']).agg({'Order ID': 'count', 'Amount': 'sum'}).reset_index()

# Top shipping methods by number of orders
fig = px.bar(shipping_data, 
             x='ship-service-level', 
             y='Order ID', 
             title="Top Shipping Methods by Number of Orders",
             labels={"Order ID": "Number of Orders"})
fig.show()
```

```python
# Top shipping methods by revenue
fig = px.bar(shipping_data, 
             x='ship-service-level', 
             y='Amount', 
             title="Top Shipping Methods by Revenue",
             labels={"Amount": "Total Sales"})
fig.show()
```
